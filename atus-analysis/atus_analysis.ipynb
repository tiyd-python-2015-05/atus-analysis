#US Time Use Data: Home Maintenance by Ownership Type

```python
rent_own_interior = group_and_average(atus, 'HETENURE', '0203') 
rent_own_exterior = group_and_average(atus, 'HETENURE', '0204') 
```

```python
N = 3
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, rent_own_interior, width, color='r', align='center')
p2 = plt.bar(ind, rent_own_exterior, width, color='y',
             bottom=rent_own_interior, align='center')

plt.ylabel('Minutes')
plt.title('Household Maintenance by Ownership Status')
plt.xticks(ind, ('Free Rent', 'Owner', 'Renter') )
# plt.yticks(np.arange(0,81,10))
plt.legend( (p1[0], p2[0]), ('Interior', 'Exterior'))

plt.show()
```

#US Time Use Data: Exercise and Health Related Care 

```python
exercise_age_group = group_and_average(atus, 'AGE_GROUPS', '1301') 
medcare_age_group = group_and_average(atus, 'AGE_GROUPS', '0804') 
personal_health = group_and_average(atus, 'AGE_GROUPS', '0103')
```

```python
N = 8
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, exercise_age_group, width, color='r', align='center')
p2 = plt.bar(ind, medcare_age_group, width, color='y',
             bottom=exercise_age_group, align='center')
p3 = plt.bar(ind, personal_health , width, color='b',
             bottom=(exercise_age_group + medcare_age_group), align='center')

plt.ylabel('Minutes')
plt.title('Exercise and Health Care')
plt.xticks(ind, ('15-20', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80+') )
# plt.yticks(np.arange(0,81,10))
plt.legend( (p1[0], p2[0], p3[0]), ('Exercise', 'Health Care', 'Health Related Self-Care'))

plt.show()
```
